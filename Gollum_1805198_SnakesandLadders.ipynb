```python
###########
##IMPORTS##
###########
from random import randint
```

```python
######################
##required variables##
######################

player_name = []
player_position = [0, 0]
snakes = {17 : 7, 54 : 34, 62 : 19, 98 : 79}
ladders = {3 : 38, 24 : 33, 42 : 93, 72 : 84}
mode = None
```

```python
#############
##functions##
#############
```

```python
def welcome():
    print('\n###### Welcome to Snakes and Ladders Game ######')
    player_name.append(input('Enter the name of player 1: '))
    player_name.append(input('Enter the name of player 2: '))
    print('There are 2 modes available: Auto and Manual')
    print('In Auto mode, player must type "roll" (without quotes) in the prompt')
    print('In Manual mode, player must enter a number between 1-19 in the prompt')
    print('Note: \n1.All inputs are case in-sensetive!\n2.At any point in the game if you want to quit, you can enter "quit" (wihtout quotes) in the prompt')
```

```python
# for checking if there is a snake on current position
def check_snake(pos):
    for current_pos, new_pos in snakes.items():
        if (current_pos == pos):
            return new_pos
    else:
        return 0
```

```python
# for checking if there is a ladder on current position
def check_ladder(pos):
    for current_pos, new_pos in ladders.items():
        if (current_pos == pos):
            return new_pos
    else:
        return 0
```

```python
# If User selects auto mode, this function will be called
def auto_mode():
    print('In Auto game mode, player must type "roll" (without quotes) in the prompt')
    print('###### Let us Start ######')
    
    current_player = 1
    while(True):
        
        if(current_player == 1):
            current_player = 1
            #player 1 goes first
            dice_val = input(f"{player_name[0]}: ").lower()
            if(dice_val == 'quit'):
                print(f'{player_name[1]} won the game!')
                return 
            if(dice_val != 'roll'):
                print("Wrong Input! Please try again.")
                continue

            rolled_val = randint(1,6)
            print(f'You got a {rolled_val}')
            if(player_position[0] + rolled_val <= 100):
                player_position[0] += rolled_val
                s = check_snake(player_position[0])
                l = check_ladder(player_position[0])
                if s!=0:
                    player_position[0] = s
                    print('You got bit by a snake! Bad luck')
                if l!=0 :
                    player_position[0] = l
                    print('You found a ladder! Nice!')

            else:
                print("Overshoot!")

            print(f'Your Final Position is: {player_position[0]}')
            if player_position[0] == 100:
                print(f"{player_name[0]} won the game!")
                return
        
        current_player = 2
        if(current_player == 2):
            #player 2's turn
            dice_val = input(f"{player_name[1]}: ").lower()
            if(dice_val == 'quit'):
                print(f'{player_name[0]} won the game!')
                return
            if(dice_val != 'roll'):
                print("Wrong Input! Please try again.")
                continue
            
            rolled_val = randint(1,6)
            print(f'You got a {rolled_val}')
            if(player_position[1] + rolled_val <= 100):
                player_position[1] += rolled_val
                s = check_snake(player_position[1])
                l = check_ladder(player_position[1])
                if s!=0:
                    player_position[1] = s
                    print('You got bit by a snake! Bad luck')
                if l!=0 :
                    player_position[1] = l
                    print('You found a ladder! Nice!')
            else:
                print("Overshoot!")

            print(f'Your Final Position is: {player_position[1]}')
            current_player = 1
            if player_position[1] == 100:
                print(f"{player_name[1]} won the game!")
                return
```

```python
#if user selects manual mode, this function will be called
def manual_mode():
    print('In Manual mode, player must enter a number between 1-19 in the prompt')
    print('###### Let us Start ######')
    
    current_player = 1
    while(True):

        if(current_player == 1):
            #player 1's turn
            dice_val = input(f"{player_name[0]}: ")
            if(dice_val.lower() == 'quit'):
                print(f'{player_name[1]} won the game!')
                return
            try:
                # dice_val = int(input(f"{player_name[0]}: "))
                dice_val = int(dice_val)
                if (dice_val in range (1,20)):
                    rolled_val = dice_val
                else:
                    print("Wrong input enter again")
                    continue
            except ValueError:
                print("Wrong input enter again")
                continue

            print(f'You got a {rolled_val}')
            if(player_position[0]+rolled_val <= 100):
                player_position[0] += rolled_val
                s = check_snake(player_position[0])
                l = check_ladder(player_position[0])
                if s!=0:
                    player_position[0] = s
                    print('You got bit by a snake! Bad luck')
                if l!=0 :
                    player_position[0] = l
                    print('You found a ladder! Nice!')
            else:
                print("Overshoot!")

            print(f'Your Final Position is: {player_position[0]}')
            if player_position[0] == 100:
                print(f"{player_name[0]} won the game!")
                return

        current_player = 2
        #player 2 
        if(current_player == 2):
            dice_val = input(f"{player_name[1]}: ")
            if(dice_val.lower() == 'quit'):
                print(f'{player_name[0]} won the game!')
                return
            try:
                dice_val = int(dice_val)
                if (dice_val in range (1,20)):
                    rolled_val = dice_val
                else:
                    print("Wrong input enter again")
                    continue
            except ValueError:
                print("Wrong input enter again")
                continue

            print(f'You got a {rolled_val}')
            if(player_position[1]+rolled_val <= 100):
                player_position[1] += rolled_val
                s = check_snake(player_position[1])
                l = check_ladder(player_position[1])
                if s!=0:
                    player_position[1] = s
                    print('You got bit by a snake! Bad luck')
                if l!=0 :
                    player_position[1] = l
                    print('You found a ladder! Nice!')
            else:
                print("Overshoot!")

            print(f'Your Final Position is: {player_position[1]}')
            current_player = 1
            if player_position[1] == 100:
                print(f"{player_name[1]} won the game!")
                return
```

```python
########
##main##
########
welcome()

while(True):
    mode = input('Please enter your choice (Auto/Manual): ').lower()
    if( mode == 'auto'):
        auto_mode()
        break
    elif( mode == 'manual' ):
        manual_mode()
        break
    else:
        print("Wrong input! Please try again")

print("###### Game Successfully Finished! ######")
```
